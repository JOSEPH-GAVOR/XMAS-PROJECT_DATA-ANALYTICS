```python
mean_FA = dataset['fixed acidity'].mean()
mean_VA = dataset['volatile acidity'].mean()
mean_CA = dataset['citric acid'].mean()
mean_RS = dataset['residual sugar'].mean()
mean_chlorides = dataset['chlorides'].mean()
mean_FSD = dataset['free sulfur dioxide'].mean()
mean_TSD = dataset['total sulfur dioxide'].mean()
mean_density = dataset['density'].mean()
mean_PH = dataset['pH'].mean()
mean_sulphates = dataset['sulphates'].mean()
mean_alcohol = dataset['alcohol'].mean()
mean_quality = dataset['quality'].mean()
```

```python
print(f'fixed acidity: {mean_FA}')
print(f'volatile acidity: {mean_VA}')
print(f'citric acid: {mean_CA}')
print(f'residual sugar: {mean_RS}')
print(f'chlorides : {mean_chlorides }')
print(f'free sulfur dioxide: {mean_FSD}')
print(f'total sulfur dioxide: {mean_TSD}')
print(f'density: {mean_density}')
print(f'PH: {mean_PH}')
print(f'sulphates: {mean_sulphates}')
print(f'alcohol: {mean_alcohol}')
print(f'quality: {mean_quality}')
```

```python
dataset
```

```python
dataset['fixed acidity'] = dataset['fixed acidity'].fillna(mean_FA)
dataset['volatile acidity'] =dataset['volatile acidity'].fillna(mean_VA)
dataset['citric acid'] =  dataset['citric acid'].fillna(mean_CA)
dataset['residual sugar'] = dataset['residual sugar'].fillna(mean_RS)
dataset['chlorides'] = dataset['chlorides'].fillna(mean_chlorides)
dataset['free sulfur dioxide'] = dataset['free sulfur dioxide'].fillna(mean_FSD)
dataset['total sulfur dioxide'] = dataset['total sulfur dioxide'].fillna(mean_TSD)
dataset['density'] = dataset['density'].fillna(mean_density)
dataset['pH'] = dataset['pH'].fillna(mean_PH)
dataset['sulphates'] = dataset['sulphates'].fillna(mean_sulphates)
dataset['alcohol'] = dataset['alcohol'].fillna(mean_alcohol )
dataset['quality'] = dataset['quality'].fillna(mean_quality)
```
